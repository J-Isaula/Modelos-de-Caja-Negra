## Función ReLU 

```python
import torch
import matplotlib.pyplot as plt

x = torch.linspace(-10,10)
relu = torch.nn.ReLU()
y = relu(x)

plt.title('ReLU')
plt.plot(x.tolist(),y.tolist())
plt.show();
```

## Función Sigmoide 

```python
import torch
import matplotlib.pyplot as plt

x = torch.linspace(-10,10)
relu = torch.nn.Sigmoid()
y = relu(x)

plt.title('Sigmoidea')
plt.plot(x.tolist(),y.tolist())
plt.show();

```

## Función Exponential LU

```python
import torch
import matplotlib.pyplot as plt

x = torch.linspace(-10,10)
relu = torch.nn.ELU()
y = relu(x)

plt.title('Exponential LU')
plt.plot(x.tolist(),y.tolist())
plt.show();
```

## Función Tanh

```python
import torch
import matplotlib.pyplot as plt

x=torch.linspace(-10,10)
tanh = torch.nn.Tanh()
y = tanh(x)

plt.title('Tanh')
plt.plot(x.tolist(),y.tolist())
plt.show()
```
